```python
import pandas as pd

df = pd.read_csv("data.csv")
df = df[df["nnodes"] == 1]
df
```

```python
from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure(dpi=150)
ax = fig.add_subplot(111)
ax.scatter(x=df["took_all"] / 60, y=df["real_time(min)"])
ax.set_xlabel("Ray Time(min)")
ax.set_ylabel("Real Time(min)")
ax.plot(
    np.linspace(20, 25, 100), np.linspace(20, 25, 100), color="red", label="y=x"
)
```
